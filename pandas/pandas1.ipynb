# Resumen descriptivo

```python
s=pd.Series([1,2,3,4,2,5,6])

print(s.count())
print(s.value_counts())
print(s.value_counts(normalize=True))
print(s.min())
print(s.max())
print(s.var())
print(s.std())
```

```python
print(s.describe())
```

```python
print(s.sum())
print(s.cumsum())
```
